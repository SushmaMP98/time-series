#time resampling

```python
df_gs.head()
```

```python
df_gs.resample(rule='A').min()# rule = 'A' means yearly end frequency
```

```python
df_gs.resample(rule='A').max()
```

```python
df_gs.resample(rule='A').max().plot()
```

```python
#quaterly start frequency
df_gs.resample(rule='QS').max()
```

```python
df_gs.resample(rule='QS').max().plot()
```

```python
#business end frequency
df_gs.resample(rule='ba').max()
```

```python
df_gs.resample(rule='ba').max().plot()
```

```python
#business quaterly frequence
df_gs.resample(rule='bqs').max().plot()
```

```python
#plotting
df_gs['GS10'].resample(rule='BA').mean().plot(kind='bar')
```

```python
#monthly frequency
df_gs['GS10'].resample(rule='m').mean()
```

```python
df_gs['GS10'].rolling(10).mean() # rolling window refers to a method of forecasting where we only use a certain amount of data at a time to make our predictions
```

```python
df_gs['GS10 20 days rolling']=df_gs['GS10'].rolling(20).mean()
```

```python
df_gs.head(21)
```

```python
df_gs[['GS10','GS10 20 days rolling']].plot(figsize=(12,5))
```
